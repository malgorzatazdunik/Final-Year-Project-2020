## Pre-processing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from bs4 import BeautifulSoup
import re
import string
import emoji
import spacy
import string

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

# Function for url's
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_mention(text):
    text = ' '.join(word for word in text.split() if word[0]!='@')
    return text

def spelling(phrase):
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
    
import re
from emot.emo_unicode import UNICODE_EMO, EMOTICONS

def replace(text):
    
    # Converting emojis to words
    def convert_emojis(text):

        tweet = emoji.demojize(text)
        tweet = tweet.replace(":"," ")
        tweet = ' '.join(tweet.split())
        
        new = list(tweet)
        for i in range(len(new)):
            if new[i] == "_":
                new[i] = " "
        return ''.join(new)
    
    smile = { ":-)": "happy", ":)" : "happy", ":-(": "sad",  ":(": "sad", ":*": "kiss", "<3": "love",
             ":D": "laugh", "D:": "sad", ":o": "surprised", ":O": "surprised", ":p": "cheeky",
             ":P": "cheeky", ">:(": "angry", ">:)": "evil", ":')" :"happy tears",
             ":'(": "sad tears"
            }

    tweet = convert_emojis(text)
    
    words = tweet.split()
    reformed = [smile[i] if i in smile else i for i in words]
    return " ".join(reformed)
```

```python
def clean_text(text):
    text = BeautifulSoup(text, "lxml").text # HTML decoding
    text = remove_urls(text)
    
    text = text.lower() # lowercase text
    text = spelling(text)
    text = replace(text)
    
    text = remove_mention(text)
    text = remove_punct(text)
    
    return text
```

```python
train_texts = []
test_texts = []
val_texts = []

for i in df.Tweet:
    train_texts.append(clean_text(i))
for i in df_test.Tweet:
    test_texts.append(clean_text(i))
for i in df_val.Tweet:
    val_texts.append(clean_text(i))
```

```python
# texts to sequences
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

maxlen = 100
max_words = 10000

tokenize = Tokenizer(num_words=max_words, char_level=False)
tokenize.fit_on_texts(train_texts) # only fit on train
word_index = tokenize.word_index

X_train = tokenize.texts_to_sequences(train_texts)
X_val = tokenize.texts_to_sequences(val_texts)
X_test = tokenize.texts_to_sequences(test_texts)

X_train = pad_sequences(X_train, maxlen=maxlen)
X_val = pad_sequences(X_val, maxlen=maxlen)
X_test = pad_sequences(X_test, maxlen=maxlen)
```

```python
glove_dir = '/Users/goszk/Desktop/labs AI/datasets/pretrained word vectors/glove.twitter.27B/glove.twitter.27B.100d.txt'
```

```python
import numpy as np

embeddings_index = {}
f = open(glove_dir, encoding = 'utf-8')
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype = 'float32')
    embeddings_index[word] = coefs
f.close()
```

```python
embedding_dim = 100
embedding_matrix = np.zeros((max_words,embedding_dim))
for word, i in word_index.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word)
        embedding_vector = embeddings_index.get(word)
        if(embedding_vector is not None):
            embedding_matrix[i] = embedding_vector
```

## Model 1.

```python
from keras.layers import Activation,LSTM, Embedding,Flatten, Dense, Bidirectional, Dropout, Conv1D, MaxPooling1D, GlobalMaxPool1D, SpatialDropout1D
from keras import metrics
from keras.models import Sequential
from keras.optimizers import Adam
from keras.losses import binary_crossentropy
from keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
from keras_self_attention import SeqSelfAttention
from keras import metrics

model1 = Sequential()
model1.add(Embedding(max_words, 100, input_length=maxlen,weights=[embedding_matrix], trainable = True ))
model1.add(SpatialDropout1D(0.5))
model1.add(Bidirectional(LSTM(units=128, return_sequences=True,
                              dropout=0.1, recurrent_dropout=0.2)))
model1.add(SeqSelfAttention(attention_activation='sigmoid'))
model1.add(Flatten())
model1.add(Dense(256, activation = 'relu'))
model1.add(Dropout(0.5))
model1.add(Dense(11, activation='sigmoid'))

model1.summary()
```

```python
from keras.utils import plot_model
plot_model(model1, to_file='graphs/EXPERIMENT_1_graph.png', show_shapes=True, show_layer_names=True)
```

```python
model1.compile(optimizer=Adam(0.001), 
               loss='binary_crossentropy', 
               metrics=['accuracy'])

callbacks = [
    #ReduceLROnPlateau(patience=5,cooldown=0, verbose=1),
    EarlyStopping(patience=10,  restore_best_weights = True),
    ModelCheckpoint(filepath='trained weights/EXPERIMENT_1.h5', save_best_only=True)
]

history = model1.fit(X_train, y_train,
                    epochs=50,
                    batch_size=256,
                    validation_data=(X_val,y_val),
                    callbacks=callbacks)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(20,20))

metrics = ['accuracy', 'loss']


for i in range(len(metrics)):
    
    ax = fig.add_subplot(4,4,i+1)
    
    ax.plot(history.history[metrics[i]])
    
    ax.plot(history.history['val_'+metrics[i]])
    
    plt.ylabel(metrics[i])
    plt.xlabel('epoch')
    plt.legend(['train','validation'])
plt.show()
```

```python
import numpy as np
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import jaccard_score, f1_score, precision_score, recall_score

def precision_recall(confusion_matrix,label):
    TN, FP, FN, TP = confusion_matrix[label].ravel()
    
    precision = TP/(TP+FP)
    recall = TP/(TP+FN)
    return precision, recall

classes = ["anger","anticipation","disgust","fear","joy","love","optimism","pessimism","sadness","surprise","trust"]

conf_matrix = multilabel_confusion_matrix(y_test, y_pred)

print("Precision ", precision_score(y_test,y_pred, average='samples', zero_division=0))
print("Recall ", recall_score(y_test,y_pred, average='samples', zero_division=0))
print("SAMPLES JACCARD SCORE ", jaccard_score(y_test,y_pred, average='samples'))
print("MACRO f1 ", f1_score(y_test, y_pred, average='macro', zero_division=0))
print("MICRO f1 ", f1_score(y_test, y_pred, average='micro', zero_division = 0), "\n")

for i in range(len(conf_matrix)):
    print(classes[i],"\n",conf_matrix[i])
 
    print("Precision: ", precision_recall(conf_matrix, i)[0], ", Recall: ", precision_recall(conf_matrix, i)[1], "\n")
```
